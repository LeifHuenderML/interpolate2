```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print('done')
```

```python
#import the datasets
corn_tx_data = pd.read_csv('../../data/agricultural_yield_datasets/USDA_NASS_Corn_Texas.csv')
corn_iowa_data = pd.read_csv('../../data/agricultural_yield_datasets/USDA_NASS_Corn_Iowa_Dataset.csv')
pecan_nm_data = pd.read_csv('../../data/agricultural_yield_datasets/USDA_NASS_Pecan_New_Mexico.csv')
```

```python
tx_n_counties = corn_tx_data["County"].value_counts()
print(f'Texas counties number of times it appears in the dataset {tx_n_counties}')
```

```python
iowa_n_counties = corn_iowa_data["County"].value_counts()
print(f'Iowa counties number of times it appears in the dataset {iowa_n_counties}')
```

```python
nm_n_counties = pecan_nm_data["County"].value_counts()
print(f'NM counties number of times it appears in the dataset {tx_n_counties}')
```

```python
#texas and iowa show more data but they also span a larger swath of time 
#this is to look at the number of years that the top 4 counties of each place covers
top5_tx = corn_tx_data[corn_tx_data['County'].isin(['JACKSON', 'MCLENNAN', 'FAYETTE', 'WILLIAMSON'])]

jackson = corn_tx_data[corn_tx_data['County'] == 'JACKSON']
mclennan = corn_tx_data[corn_tx_data['County'] == 'MCLENNAN']
fayette = corn_tx_data[corn_tx_data['County'] == 'FAYETTE']
williamson = corn_tx_data[corn_tx_data['County'] == 'WILLIAMSON']

plt.plot(jackson['Year'], jackson['Value'], label = "jackson")
plt.plot(mclennan['Year'], mclennan['Value'], label = "mclennan")
plt.plot(fayette['Year'], fayette['Value'], label = "fayette")
plt.plot(williamson['Year'], williamson['Value'], label = "williamson")

plt.legend()
plt.xlabel('Years')
plt.ylabel('Yield')
plt.title('Top 4 Most Occuring Counties in Texas')

plt.show()
```

```python
top5_iowa = corn_iowa_data[corn_iowa_data['County'].isin(['PALO ALTO', 'BREMER', 'CLAYTON', 'DELAWARE'])]

PALO_ALTO = corn_iowa_data[corn_iowa_data['County'] == 'PALO ALTO']
BREMER = corn_iowa_data[corn_iowa_data['County'] == 'BREMER']
CLAYTON = corn_iowa_data[corn_iowa_data['County'] == 'CLAYTON']
DELAWARE = corn_iowa_data[corn_iowa_data['County'] == 'DELAWARE']

plt.plot(PALO_ALTO['Year'], PALO_ALTO['Value'], label = "PALO ALTO")
plt.plot(BREMER['Year'], BREMER['Value'], label = "BREMER")
plt.plot(CLAYTON['Year'], CLAYTON['Value'], label = "CLAYTON")
plt.plot(DELAWARE['Year'], DELAWARE['Value'], label = "DELAWARE")

plt.legend()
plt.xlabel('Years')
plt.ylabel('Yield')
plt.title('Top 4 Most Occuring Counties in Texas')

plt.show()
```
