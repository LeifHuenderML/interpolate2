```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print('done')
```

```python
#import the datasets
New_Mexico_dataset = pd.read_csv('../../data/agricultural_yield_datasets/New_Mexico_Pecan_Yield.csv')
```

```python
nm_n_counties = New_Mexico_dataset["County"].value_counts()
print(f'NM counties number of times it appears in the dataset {nm_n_counties}')
```

```python

top5_nm = New_Mexico_dataset[New_Mexico_dataset['County'].isin(['DONA ANA', 'CHAVES', 'SIERRA', 'OTERO'])]

DONA_ANA = New_Mexico_dataset[New_Mexico_dataset['County'] == 'DONA ANA']
CHAVES = New_Mexico_dataset[New_Mexico_dataset['County'] == 'CHAVES']
SIERRA = New_Mexico_dataset[New_Mexico_dataset['County'] == 'SIERRA']
OTERO = New_Mexico_dataset[New_Mexico_dataset['County'] == 'OTERO']

plt.plot(DONA_ANA['Year'], DONA_ANA['Value'], label = "DONA ANA")
plt.plot(CHAVES['Year'], CHAVES['Value'], label = "CHAVES")
plt.plot(SIERRA['Year'], SIERRA['Value'], label = "SIERRA")
plt.plot(OTERO['Year'], OTERO['Value'], label = "OTERO")

plt.legend()
plt.xlabel('Years')
plt.ylabel('Yield')
plt.title('Top 4 Most Occuring Counties in New Mexico')

plt.show()
```
