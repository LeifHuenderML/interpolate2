```python
import pandas as pd

def filter_and_group(df, start_year, end_year, months):
    """
    Filter and group a DataFrame based on specified year range and months.

    Parameters:
    - df: Pandas DataFrame with a datetime column named 'date'.
    - start_year: int, the starting year for the filter.
    - end_year: int, the ending year for the filter.
    - months: list of int, months to include in the filter (1=Jan, 2=Feb, ..., 12=Dec).

    Returns:
    - A new DataFrame filtered by the specified years and months, grouped by each period.
    """

    # Ensure the 'date' column is a datetime type
    if not pd.api.types.is_datetime64_any_dtype(df['date']):
        df['date'] = pd.to_datetime(df['date'])

    # Filter by year and month
    mask = (df['date'].dt.year >= start_year) & (df['date'].dt.year <= end_year) & (df['date'].dt.month.isin(months))
    filtered_df = df.loc[mask]

    # Group by year and month, assuming we just need to see grouped data
    # If you need to perform any aggregation, you can modify this part
    grouped_df = filtered_df.groupby([filtered_df['date'].dt.year, filtered_df['date'].dt.month])

    # If you want to do operations within groups or visualize them separately:
    # for (year, month), group in grouped_df:
    #     print(f"Data for {year}-{month}:")
    #     print(group)

    return grouped_df

# Example usage:
# Assuming 'data' is your DataFrame and it contains a 'date' column with dates ranging from 1970 to 2000
# data = pd.DataFrame(...)
# result = filter_and_group(data, 1995, 1999, [1, 2, 3, 4])
# Now 'result' holds the grouped data
```
