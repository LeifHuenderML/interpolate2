```python
weather_data = pd.read_csv('../../data/weather_data/weather_data copy.csv')
weather_data.head()
```

```python
# relabel the missing locations 

# Update the 'city_name' for rows where 'lat' is 28.581 and 'lon' is 77.173
weather_data.loc[(weather_data['lat'] == 28.581) & (weather_data['lon'] == 77.173), 'city_name'] = 'Northern New Delhi'

# Update the 'city_name' for rows where 'lat' is 28.5615 and 'lon' is 77.1935
weather_data.loc[(weather_data['lat'] == 28.5615) & (weather_data['lon'] == 77.1935), 'city_name'] = 'Southern New Delhi'

# Update the 'city_name' for rows where 'lat' is 22.69 and 'lon' is 72.856
weather_data.loc[(weather_data['lat'] == 22.69) & (weather_data['lon'] == 72.856), 'city_name'] = 'Gujarat'

# Update the 'city_name' for rows where 'lat' is 11.2962 and 'lon' is 76.9414
weather_data.loc[(weather_data['lat'] == 11.2962) & (weather_data['lon'] == 76.9414), 'city_name'] = 'Tamil Nadu'
```

```python
all_nan_columns = weather_data.isna().all()
print(all_nan_columns[all_nan_columns].index.tolist())

weather_data = weather_data.dropna(axis=1, how ='all')
```

```python
#remove majorily missing data 
nan_prop = weather_data.isna().mean()
threshold = 0.6
majority_nan_columns = nan_prop[nan_prop > threshold].index.to_list()
print(majority_nan_columns)
```

```python
weather_data =  weather_data.drop(['visibility', 'wind_gust', 'rain_1h', 'rain_3h', 'snow_1h', 'snow_3h', 'dt_iso', 'timezone', 'weather_icon'], axis=1)
```

```python
# Load the datasets
df = pd.read_csv('../../data/weather_data/labelled_weather_data.csv')

# Split the dataset into individual streams for pecans
dona_ana_df = df[df['city_name'] == 'Doña Ana']
chaves_df = df[df['city_name'] == 'Chaves County']
sierra_df = df[df['city_name'] == 'Sierra County']
otero_df = df[df['city_name'] == 'Otero County']

# Split the dataset into individual streams for malaria outbreaks
north_new_delhi_df = df[df['city_name'] == 'Northern New Delhi']
south_new_delhi_df = df[df['city_name'] == 'Southern New Delhi']
gujarat_df = df[df['city_name'] == 'Gujarat']
tamil_nadu_df = df[df['city_name'] == 'Tamil Nadu']


print(f'Dona Ana Shape: {dona_ana_df.shape}')
print(f'Chaves Shape: {chaves_df.shape}')
print(f'Otero Shape: {otero_df.shape}')
print(f'Sierra Shape: {sierra_df.shape}')
print()
print(f'North New Delhi Shape: {north_new_delhi_df.shape}')
print(f'South New Delhi Shape: {south_new_delhi_df.shape}')
print(f'Gujarat Shape: {gujarat_df.shape}')
print(f'Tamil Nadu Shape: {tamil_nadu_df.shape}')

#the shape is off for all of them so we need to alignt them via their timespamps so they will have accurat scoring
```

```python
#convert them all from unix seconds to pd's datetime format

dona_ana_df.loc[:, 'dt'] = pd.to_datetime(dona_ana_df['dt'], unit='s', utc=True)
chaves_df.loc[:, 'dt'] = pd.to_datetime(chaves_df['dt'], unit='s', utc=True)
sierra_df.loc[:, 'dt'] = pd.to_datetime(sierra_df['dt'], unit='s', utc=True)
otero_df.loc[:, 'dt'] = pd.to_datetime(otero_df['dt'], unit='s', utc=True)

north_new_delhi_df.loc[:, 'dt'] = pd.to_datetime(north_new_delhi_df['dt'], unit='s', utc=True)
south_new_delhi_df.loc[:, 'dt'] = pd.to_datetime(south_new_delhi_df['dt'], unit='s', utc=True)
gujarat_df.loc[:, 'dt'] = pd.to_datetime(gujarat_df['dt'], unit='s', utc=True)
tamil_nadu_df.loc[:, 'dt'] = pd.to_datetime(tamil_nadu_df['dt'], unit='s', utc=True)

# get rid of duplicates that will misalign timestamps
for df in [dona_ana_df, chaves_df, otero_df, sierra_df]:
    print(df.duplicated('dt').sum())

dona_ana_df = dona_ana_df.drop_duplicates('dt')
chaves_df = chaves_df.drop_duplicates('dt')
otero_df = otero_df.drop_duplicates('dt')
sierra_df = sierra_df.drop_duplicates('dt')

print()
for df in [north_new_delhi_df, south_new_delhi_df, gujarat_df, tamil_nadu_df]:
    print(df.duplicated('dt').sum())

north_new_delhi_df = north_new_delhi_df.drop_duplicates('dt')
south_new_delhi_df = south_new_delhi_df.drop_duplicates('dt')
gujarat_df = gujarat_df.drop_duplicates('dt')
tamil_nadu_df = tamil_nadu_df.drop_duplicates('dt')

#find the common dt accross all data sets
common_dt = pd.Series(list(set(dona_ana_df['dt']) & set(chaves_df['dt']) & set(otero_df['dt']) & set(sierra_df['dt'])))

#align the datasets by datetime
dona_ana_df = dona_ana_df[dona_ana_df['dt'].isin(common_dt)]
chaves_df = chaves_df[chaves_df['dt'].isin(common_dt)]
otero_df = otero_df[otero_df['dt'].isin(common_dt)]
sierra_df = sierra_df[sierra_df['dt'].isin(common_dt)]

#find the common dt accross all data sets
common_dt = pd.Series(list(set(north_new_delhi_df['dt']) & set(south_new_delhi_df['dt']) & set(gujarat_df['dt']) & set(tamil_nadu_df['dt'])))

#align the datasets by datetime
north_new_delhi_df = north_new_delhi_df[north_new_delhi_df['dt'].isin(common_dt)]
south_new_delhi_df = south_new_delhi_df[south_new_delhi_df['dt'].isin(common_dt)]
gujarat_df = gujarat_df[gujarat_df['dt'].isin(common_dt)]
tamil_nadu_df = tamil_nadu_df[tamil_nadu_df['dt'].isin(common_dt)]

print(f'Dona Ana Shape: {dona_ana_df.shape}')
print(f'Chaves Shape: {chaves_df.shape}')
print(f'Otero Shape: {otero_df.shape}')
print(f'Sierra Shape: {sierra_df.shape}')
print()
print(f'North New Delhi Shape: {north_new_delhi_df.shape}')
print(f'South New Delhi Shape: {south_new_delhi_df.shape}')
print(f'Gujarat Shape: {gujarat_df.shape}')
print(f'Tamil Nadu Shape: {tamil_nadu_df.shape}')
```

```python
#save each as a cleaned independant stream

dona_ana_df.to_csv('../../data/weather_data/cleaned_data/cleaned_dona_ana_df', index=False)
chaves_df.to_csv('../../data/weather_data/cleaned_data/cleaned_chaves_df', index=False)
otero_df.to_csv('../../data/weather_data/cleaned_data/cleaned_otero_df', index=False)
sierra_df.to_csv('../../data/weather_data/cleaned_data/cleaned_sierra_df',index=False)

north_new_delhi_df.to_csv('../../data/weather_data/cleaned_data/cleaned_north_new_delhi_df', index=False)
south_new_delhi_df.to_csv('../../data/weather_data/cleaned_data/cleaned_south_new_delhi_df', index=False)
gujarat_df.to_csv('../../data/weather_data/cleaned_data/cleaned_gujarat_df', index=False)
tamil_nadu_df.to_csv('../../data/weather_data/cleaned_data/cleaned_tamil_nadu_df', index=False)
```

```python
#join them all together and save as group straems
pecan_df = pd.concat([dona_ana_df, chaves_df, otero_df, sierra_df] , ignore_index=True)

malaria_df = pd.concat([north_new_delhi_df, south_new_delhi_df, gujarat_df, tamil_nadu_df], ignore_index=True)

pecan_df.to_csv('../../data/weather_data/cleaned_data/cleaned_pecan_weather_dataset')
malaria_df.to_csv('../../data/weather_data/cleaned_data/cleaned_malaria_weather_dataset')
```
